```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('tdw').getOrCreate()
```

```python
data = spark.read.csv('ks-projects-201612.csv', inferSchema=True, header=True)
```

```python
data.printSchema()
```

```python
data.show(5)
```

```python
country_project = data.groupBy('country ').count()
```

```python
country_project.orderBy('country ').show(1000)
```

```python
country_project.where('count > 1000').show()
```
